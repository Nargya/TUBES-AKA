```python
from prettytable import PrettyTable
import time
import matplotlib.pyplot as plt
```

```python
# Fungsi untuk menghitung permutasi secara rekursif
def permutasi_rekursif(n, r):
    if r == 0:
        return 1
    else:
        return n * permutasi_rekursif(n - 1, r - 1)

# Fungsi untuk menghitung permutasi secara iteratif
def permutasi_iteratif(n, r):
    result = 1
    for i in range(r):
        result *= n - i
    return result
```

```python
# Fungsi untuk menjalankan program dan mengumpulkan data
def main():
    table = PrettyTable()
    table.field_names = ["n (Jumlah Karyawan)", "Waktu Iteratif (detik)", "Waktu Rekursif (detik)"]

    n_values = []
    iter_times = []
    recur_times = []

    while True:
        # Meminta input dari pengguna
        n = int(input("Masukkan jumlah karyawan (n) atau -1 untuk keluar: "))
        if n == -1:
            break

        r = 4  # Nilai r sudah tetap 4 sesuai dengan studi kasus

        # Mengukur waktu untuk permutasi iteratif
        start_time = time.time()
        perm_iter = permutasi_iteratif(n, r)
        iter_time = time.time() - start_time

        # Mengukur waktu untuk permutasi rekursif
        start_time = time.time()
        perm_recur = permutasi_rekursif(n, r)
        recur_time = time.time() - start_time

        # Menambahkan data ke dalam tabel dengan format 5 angka di belakang koma
        table.add_row([n, f"{iter_time:.5e}", f"{recur_time:.5e}"])

        # Menyimpan data untuk grafik
        n_values.append(n)
        iter_times.append(iter_time)
        recur_times.append(recur_time)

    # Menampilkan tabel PrettyTable
    print(table)

    # Membuat grafik perbandingan waktu
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, iter_times, label='Iteratif', color='b', marker='o')
    plt.plot(n_values, recur_times, label='Rekursif', color='r', marker='x')
    plt.xlabel("Jumlah Karyawan (n)")
    plt.ylabel("Waktu Eksekusi (detik)")
    plt.title("Perbandingan Waktu Eksekusi Algoritma Iteratif vs Rekursif")
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
  main()

```
